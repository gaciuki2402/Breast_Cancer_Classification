```python
from google.colab import drive
drive.mount("/content/drive")
# importing the dependancies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the dataset from sklearn

breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# printing the first five rows of the dataframe
data_frame.head()
```

```python
# adding the target column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# printing the last 5 rows of the data frame
data_frame.tail()
```

```python
# nmber of rows and columns in the dataset
data_frame.shape
```

```python
# getting some info about the data
data_frame.info()
```

```python
# checking for any missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of target variables
data_frame['label'].value_counts()
```

```python
print("1 represents Benign")
print("0 represents Malignant")
```

```python
# getting the mean of each of the labels
data_frame.groupby('label').mean()
```

```python
# separating the features and the targets
X = data_frame.drop(columns = 'label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# buliding the neural network
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of the neural network
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20,activation='relu'),
                          keras.layers.Dense(2,activation='sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
              )
```

```python
# training the neural network
# epochs is the number of times the model haas to go through the data
# history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
# standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
# visualizing accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')

plt.legend(['training data','validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.xlabel('epoch')
plt.ylabel('loss')

plt.legend(['training data','validation data'], loc = 'upper right')
```

```python
# accuracy of the model on the data
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
